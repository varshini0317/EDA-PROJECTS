```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Air Polluant.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
print("Missing Values Before Handling:")
print(df.isnull().sum())
```

```python
df.replace('NA', pd.NA, inplace=True)
df['pollutant_min'] = pd.to_numeric(df['pollutant_min'], errors='coerce')
df['pollutant_max'] = pd.to_numeric(df['pollutant_max'], errors='coerce')
df['pollutant_avg'] = pd.to_numeric(df['pollutant_avg'], errors='coerce')
```

```python
# Impute missing numerical values with the mean
df.fillna(df.mean(), inplace=True)
```

```python
# Check for missing values after handling
print("\nMissing Values After Handling:")
print(df.isnull().sum())
```

```python

# Convert 'last_update' column to datetime format
df['last_update'] = pd.to_datetime(df['last_update'], format='%m-%d-%Y %H:%M:%S')

```

```python
# Print the DataFrame to verify changes
print("\nDataFrame after handling missing values and converting 'last_update':")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of Pollutants - Boxplot
# Set the style for better aesthetics
sns.set(style="whitegrid")

# Create a color palette for better differentiation
palette = sns.color_palette("Set2")
```

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x='pollutant_id', y='pollutant_avg', data=df, palette=palette)
sns.swarmplot(x='pollutant_id', y='pollutant_avg', data=df, color='0.25')  # Adding individual data points

plt.title('Distribution of Pollutants')
plt.xlabel('Pollutant ID')
plt.ylabel('Pollutant Average')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add a legend for better interpretation
plt.legend(labels=['Data Points'], loc='upper right')

plt.show()
```

```python
# Trends over Time - Line Plot
# Create a line plot with markers and a legend
plt.figure(figsize=(14, 10))
sns.lineplot(x='last_update', y='pollutant_avg', hue='city', data=df, marker='o', markersize=8, linewidth=2)
plt.title('Air Quality Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Pollutant Average')

# Format the x-axis dates for better readability
plt.xticks(rotation=45, ha='right')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=10))  # Adjust the number of x-axis ticks

# Move the legend to a more appropriate position
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import folium
```

```python
# Geographical Map - Folium
# Create a map centered around the mean latitude and longitude
map_center = [df['latitude'].mean(), df['longitude'].mean()]
folium_map = folium.Map(location=map_center, zoom_start=6)

# Add markers for each station
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['longitude']],
                  popup=f"{row['city']} - {row['station']}",
                  icon=folium.Icon(color='blue')).add_to(folium_map)

# Display the map
folium_map.save('pollution_map.html')  # Save the map as an HTML file
```

```python
#Heatmap
plt.figure(figsize=(12, 8))
heatmap_data = df.pivot_table(index='city', columns='pollutant_id', values='pollutant_avg', aggfunc='mean')
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.title('Pollutant Average Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Advanced Visualization - 3D Plot
fig = px.scatter_3d(df, x='latitude', y='longitude', z='pollutant_avg',
                    color='pollutant_avg', size='pollutant_avg',
                    hover_data=['city', 'station'],
                    title='3D Plot of Pollutant Averages by Location')
fig.show()
```

```python
# Correlation Analysis - Heatmap
correlation_matrix = df[['pollutant_min', 'pollutant_max', 'pollutant_avg', 'latitude', 'longitude']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
